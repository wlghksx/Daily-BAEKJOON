# 10250번 ACM 호텔

```python
n = int(input())
for i in range(n):
  floor,room,cust = map(int,input().split())
  a = cust%floor
  b = cust//floor
  if a==0 :
    a = floor
    b -= 1
  print(a*100+(b+1))
```

# 2755번 부녀회장이 될테야

```python
n = int(input())
for i in range(n) :
  k = int(input())
  n = int(input())
  _floor = list(range(1,n+1))
  for j in range(k):
    for k in range(1,n):
      _floor[k] += _floor[k-1]
  print(_floor[-1])
    
```
