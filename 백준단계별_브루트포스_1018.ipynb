# 1018 체스판 다시 칠하기

image.png

```python
row, col = map(int,input().split())

arr_input = []

for i in range(row):
  arr_input.append(input())

count = []

for i  in range(row-7):
  for j in range(col-7):
    count_w = 0
    count_b = 0
    for a in range(i,i+8):
      for b in range(j,j+8):
        if (a+b) % 2 == 0:
          if arr_input[a][b] != 'W': count_w += 1
          if arr_input[a][b] != 'B': count_b+= 1
        else:
          if arr_input[a][b] != 'B': count_w += 1
          if arr_input[a][b] != 'W': count_b += 1
    count.append(count_w)
    count.append(count_b)
    

print(min(count))
```
